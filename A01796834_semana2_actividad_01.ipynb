*   **Pregunta 1.** 

Busca y elimina todos los saltos de línea '\n' que se encuentran al final de cada comentario. 

Una vez finalizado, imprime los primeros 10 comentarios del resultado obtenido.


```python
p1 = [re.sub(r'$\n',r'', line) for line in docs]
```

*   **Pregunta 2.**  

Busca e imprime todas las palabras que terminan con dos o más signos de admiración seguidos, por ejemplo "!!!". 

Debes imprimir tanto la palabra como la totalidad de signos de admiración que le siguen. 

Indica cuántos resultados obtuviste.



```python
# Expresion regular que identifica las palabras que terminan con 2 o mas signos de admiración

p2 = [word for line in docs_n for word in re.findall(r'\b\w+!{2,}', line)]
```

```python
p2
```

```python
# Cuantos resultados se obtuvo

print("En la variable docs, se tienen {} resultados que cumplen con que su terminación son dos o mas signos de admiración !".format(len(p2)))
```

*   **Pregunta 3.**  

Busca e imprime todas las palabras que están escritas totalmente en mayúsculas. Cada coincidencia debe ser una sola palabra.

Indica cuántas palabras encontraste.



```python
# Expresion regular que identifica las palabras totalmente en mayusculas

# Caso 1: Se cuentan aquellas palabras de una sola letra como I o A.
p3_a = [word for line in docs_n for word in re.findall(r'\b[A-Z]+\b', line)]

# Caso 2: No se cuentan aquellas palabras de una sola letra como I o A.
p3_b = [word for line in docs_n for word in re.findall(r'\b[A-Z]+\b', line) if len(word) > 1 ]
```

```python
# Imprimimos el Caso 1
print(p3_a)
```

```python
# Imprimimos el Caso 2
print(p3_b)
```

```python
# Total de palabras totalmente en mayusculas

print("En la variable docs, se tienen {} palabras que cumplen estar totalmente en MAYUS considerando las palabras de 1 letra o mas".format(len(p3_a)))

print("En la variable docs, se tienen {} palabras que cumplen estar totalmente en MAYUS y no se consideran a las palabras de 1 sola letra".format(len(p3_b)))
```

*   **Pregunta 4.**  

Busca e imprime los comentarios en donde todos los caracteres alfabéticos (letras) están en mayúsculas. 

Cada coincidencia encontrada debe ser todo el comentario/enunciado.

Indica cuántos resultados obtuviste.


```python
# Comentarios donde todos los caracteres son alfabeticos son mayúsculas

p4 = [line for line in docs_n if re.match(r'^(?=.*[A-Z])[^a-z]+$', line)]
```

```python
print(p4)
```

```python
# Todo el comentario Alfanumérico esta totalmente en MAYUS

print("En la variable docs, se tienen {} coincidencias donde TODOS los caracteres alfanuméricos estan en mayusculas".format(len(p4)))
```

#### Como nota, la pregunta dice que todas las letras deben ser mayusculas, no se menciona en ningun momento que no pueden existir otros elementos no alfanumericos como & o !

*   **Pregunta 5.**  

Busca e imprime todas las palabras que tengan una vocal acentuada, del tipo á, é, í, ó, ú. 

Indica cuántos resultados obtuviste.

*   **Pregunta 7.**

*   **Pregunta 8.**

*   **Pregunta 12.**  

Busca e imprime todas las palabras que terminan en "ing" o "ed". 

Indica la cantidad de palabras que encontraste de cada una.

```python
p12 = [word for line in docs for word in re.findall(r'(?i)\w+(?:ing|ed)\b', line)]
```

# **Parte 3. Proceso de limpieza.**

*   **Pregunta 13.**  

Ahora realiza un proceso de limpieza del corpus que incluya los siguientes procesos:

*   Solo se deben considerar caracteres alfabéticos. Es decir, se eliminan todos los signos de puntuación y caracteres especiales.
*   Todos los caracteres alfabéticos se transforman a minúsculas.
*   Se deben eliminar todos los espacios en blanco adicionales que se puedan encontrar en cada comentario.

Al finalizar dicho proceso de limpieza, imprime el resultado de los primeros 10 comentarios resultantes. 
   




```python
p13 = [' '.join(re.sub(r'[^a-zA-ZáéíóúÁÉÍÓÚ\s]', '', line).lower().split()) for line in docs_n]
```

```python
p13[:10]
```

*   **Pregunta 14.**  

Con el resultado de la limpieza obtenido en la pregunta anterior, realiza ahora un proceso de tokenización por palabras del corpus. 

Es decir, al final de este proceso de tokenización, debes tener como resultado una lista de listas, donde cada comentario estará tokenizado por palabras.

Al terminar calcula el total de tokens obtenido en todo el corpus. 

```python
p14 = [line.split() for line in p13]
```

```python
print(p14[:10])
```

```python
n_tokens = sum(len(line) for line in p14)
print("El numero total de Tokens obtenido en todo el corpus es {}".format(n_tokens))
```
